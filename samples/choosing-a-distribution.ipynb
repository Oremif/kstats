# Choosing a Distribution

Every distribution encodes assumptions about what values are possible and how likely they are.
This notebook groups distributions by the kind of data they model and shows how to verify the fit.

| Data type | Example domain | Distribution | Constructor |
| --- | --- | --- | --- |
| Symmetric measurements around a center | User session duration | `NormalDistribution` | `NormalDistribution(mu, sigma)` |
| Heavier tails than Normal, small samples | Estimated means with few observations | `StudentTDistribution` | `StudentTDistribution(degreesOfFreedom)` |
| Positive waiting times or durations | Time between server errors | `ExponentialDistribution` | `ExponentialDistribution(rate)` |
| Positive durations with wear-out pattern | Hardware component lifetime | `WeibullDistribution` | `WeibullDistribution(shape, scale)` |
| Right-skewed positive values | API response times | `LogNormalDistribution` | `LogNormalDistribution(mu, sigma)` |
| Proportions or rates in [0, 1] | Click-through rate | `BetaDistribution` | `BetaDistribution(alpha, beta)` |
| Event counts per interval | Errors per hour | `PoissonDistribution` | `PoissonDistribution(rate)` |
| Successes in fixed trials | Conversions out of page views | `BinomialDistribution` | `BinomialDistribution(trials, probability)` |
| Overdispersed counts | Support tickets per day | `NegativeBinomialDistribution` | `NegativeBinomialDistribution(successes, probability)` |
| Heavy right tail on positive data | Income distribution, file sizes | `ParetoDistribution` | `ParetoDistribution(shape, scale)` |

## Durations and Waiting Times

### Response Times (LogNormal)

```kotlin
val responseTime = LogNormalDistribution(mu = 4.5, sigma = 0.8)

println("Mean response time: ${responseTime.mean} ms")
println("P95 latency: ${responseTime.quantile(0.95)} ms")
println("P99 latency: ${responseTime.quantile(0.99)} ms")
println("P(< 200ms): ${responseTime.cdf(200.0)}")
```

```kotlin
val xLN = (1..300).map { it.toDouble() }
val yLN = xLN.map { responseTime.pdf(it) }
val p95 = responseTime.quantile(0.95)
val p99 = responseTime.quantile(0.99)

plot {
    area {
        x(xLN); y(yLN)
        fillColor = Color.hex("#4dabf7")
        alpha = 0.4
    }
    line {
        x(xLN); y(yLN)
        color = Color.hex("#1971c2")
        width = 1.5
    }
    line {
        x(listOf(p95, p95)); y(listOf(0.0, responseTime.pdf(p95)))
        color = Color.hex("#e03131")
        width = 1.5
    }
    line {
        x(listOf(p99, p99)); y(listOf(0.0, responseTime.pdf(p99)))
        color = Color.hex("#c2255c")
        width = 1.5
    }
    layout {
        title = "LogNormal PDF — API Response Times (P95=${p95.toInt()} ms, P99=${p99.toInt()} ms)"
        size = 800 to 400
    }
}
```

```kotlin
val yCDF = xLN.map { responseTime.cdf(it) }

plot {
    line {
        x(xLN); y(yCDF)
        color = Color.hex("#1971c2")
        width = 2.0
    }
    line {
        x(listOf(0.0, 300.0)); y(listOf(0.95, 0.95))
        color = Color.hex("#e03131")
        width = 1.0
    }
    line {
        x(listOf(0.0, 300.0)); y(listOf(0.99, 0.99))
        color = Color.hex("#c2255c")
        width = 1.0
    }
    layout {
        title = "LogNormal CDF — Cumulative Probability"
        size = 800 to 400
    }
}
```

### Time Between Events (Exponential)

```kotlin
val timeBetweenErrors = ExponentialDistribution(rate = 0.5)

println("Average time between errors: ${timeBetweenErrors.mean}")
println("P(wait > 10): ${timeBetweenErrors.sf(10.0)}")
println("Median waiting time: ${timeBetweenErrors.quantile(0.5)}")
```

```kotlin
val xExp = (0..150).map { it / 10.0 }
val yExpPdf = xExp.map { timeBetweenErrors.pdf(it) }
val yExpSf = xExp.map { timeBetweenErrors.sf(it) }

plot {
    area {
        x(xExp); y(yExpPdf)
        fillColor = Color.hex("#69db7c")
        alpha = 0.4
    }
    line {
        x(xExp); y(yExpPdf)
        color = Color.hex("#2b8a3e")
        width = 2.0
    }
    line {
        x(xExp); y(yExpSf)
        color = Color.hex("#e03131")
        width = 2.0
    }
    layout {
        title = "Exponential: PDF (green) and Survival Function (red)"
        size = 800 to 400
    }
}
```

### Component Lifetime (Weibull)

```kotlin
val componentLifetime = WeibullDistribution(shape = 2.0, scale = 5000.0)

println("10% fail before: ${componentLifetime.quantile(0.1)} hours")
println("P(fail < 3000h): ${componentLifetime.cdf(3000.0)}")
println("P(survive > 4000h): ${componentLifetime.sf(4000.0)}")
```

```kotlin
val xW = (0..100).map { it * 100.0 }
val weibull05 = WeibullDistribution(shape = 0.5, scale = 5000.0)
val weibull10 = WeibullDistribution(shape = 1.0, scale = 5000.0)
val weibull20 = WeibullDistribution(shape = 2.0, scale = 5000.0)
val weibull35 = WeibullDistribution(shape = 3.5, scale = 5000.0)

val allX = xW + xW + xW + xW
val allY = xW.map { weibull05.pdf(it) } + xW.map { weibull10.pdf(it) } + xW.map { weibull20.pdf(it) } + xW.map { weibull35.pdf(it) }
val allShape = xW.map { "shape=0.5 (burn-in)" } + xW.map { "shape=1.0 (constant)" } + xW.map { "shape=2.0 (wear-out)" } + xW.map { "shape=3.5 (rapid wear)" }

plot {
    line {
        x(allX); y(allY)
        color(allShape)
        width = 2.0
    }
    layout {
        title = "Weibull PDF — Effect of Shape Parameter"
        size = 800 to 400
    }
}
```

```kotlin
val ySF = xW.map { componentLifetime.sf(it) }

plot {
    line {
        x(xW); y(ySF)
        color = Color.hex("#1971c2")
        width = 2.0
    }
    line {
        x(listOf(3000.0, 3000.0)); y(listOf(0.0, 1.0))
        color = Color.hex("#e03131")
        width = 1.0
    }
    line {
        x(listOf(4000.0, 4000.0)); y(listOf(0.0, 1.0))
        color = Color.hex("#f08c00")
        width = 1.0
    }
    layout {
        title = "Weibull Survival Function (shape=2.0, scale=5000)"
        size = 800 to 400
    }
}
```

## Counts and Events

### Event Counts (Poisson)

```kotlin
val errorsPerHour = PoissonDistribution(rate = 3.2)

println("P(0 errors): ${errorsPerHour.pmf(0)}")
println("P(<=5 errors): ${errorsPerHour.cdf(5)}")
println("99th percentile: ${errorsPerHour.quantileInt(0.99)} errors")
```

```kotlin
val kValues = (0..12).toList()
val pmfValues = kValues.map { errorsPerHour.pmf(it) }

plot {
    bars {
        x(kValues); y(pmfValues)
        fillColor = Color.hex("#9775fa")
        alpha = 0.8
    }
    layout {
        title = "Poisson PMF — Errors per Hour (lambda=3.2)"
        size = 800 to 400
    }
}
```

```kotlin
val cdfValues = kValues.map { errorsPerHour.cdf(it) }

plot {
    line {
        x(kValues.map { it.toDouble() }); y(cdfValues)
        color = Color.hex("#7048e8")
        width = 2.0
    }
    points {
        x(kValues.map { it.toDouble() }); y(cdfValues)
        color = Color.hex("#7048e8")
        size = 4.0
    }
    line {
        x(listOf(0.0, 12.0)); y(listOf(0.99, 0.99))
        color = Color.hex("#e03131")
        width = 1.0
    }
    layout {
        title = "Poisson CDF — Cumulative Probability"
        size = 800 to 400
    }
}
```

### Fixed Trials (Binomial)

```kotlin
val conversions = BinomialDistribution(trials = 1000, probability = 0.035)

println("Expected conversions: ${conversions.mean}")
println("P(exactly 40): ${conversions.pmf(40)}")
println("P(more than 50): ${conversions.sf(50)}")
```

```kotlin
val bkValues = (15..60).toList()
val bpmfValues = bkValues.map { conversions.pmf(it) }

plot {
    bars {
        x(bkValues); y(bpmfValues)
        fillColor = Color.hex("#20c997")
        alpha = 0.8
    }
    layout {
        title = "Binomial PMF — Conversions (n=1000, p=0.035)"
        size = 800 to 400
    }
}
```

### Overdispersed Counts (Negative Binomial)

```kotlin
val tickets = NegativeBinomialDistribution(successes = 5, probability = 0.1)

println("Expected ticket count: ${tickets.mean}")
println("Variance: ${tickets.variance} (> mean = overdispersion)")
```

```kotlin
val compK = (0..120).toList()
val poissonComp = PoissonDistribution(rate = 45.0) // same mean as NegBin
val poisPmf = compK.map { poissonComp.pmf(it) }
val nbPmf = compK.map { tickets.pmf(it) }

val allKComp = compK + compK
val allPmfComp = poisPmf + nbPmf
val allDist = compK.map { "Poisson(lambda=45)" } + compK.map { "NegBinomial(r=5, p=0.1)" }

plot {
    line {
        x(allKComp.map { it.toDouble() }); y(allPmfComp)
        color(allDist)
        width = 2.0
    }
    layout {
        title = "Poisson vs Negative Binomial — Same Mean, Different Spread"
        size = 800 to 400
    }
}
```

## Proportions and Rates

```kotlin
val ctr = BetaDistribution(alpha = 120.0, beta = 3880.0)

println("Point estimate of CTR: ${ctr.mean}")
println("95% credible interval: [${ctr.quantile(0.025)}, ${ctr.quantile(0.975)}]")
println("P(CTR < 3.5%): ${ctr.cdf(0.035)}")
```

```kotlin
val xBeta = (0..600).map { it / 10000.0 } // 0 to 0.06
val yBeta = xBeta.map { ctr.pdf(it) }
val ci025 = ctr.quantile(0.025)
val ci975 = ctr.quantile(0.975)

plot {
    area {
        x(xBeta); y(yBeta)
        fillColor = Color.hex("#ffd43b")
        alpha = 0.3
    }
    line {
        x(xBeta); y(yBeta)
        color = Color.hex("#f08c00")
        width = 2.0
    }
    line {
        x(listOf(ci025, ci025)); y(listOf(0.0, yBeta.max()!!))
        color = Color.hex("#e03131")
        width = 1.5
    }
    line {
        x(listOf(ci975, ci975)); y(listOf(0.0, yBeta.max()!!))
        color = Color.hex("#e03131")
        width = 1.5
    }
    layout {
        title = "Beta PDF — Click-Through Rate (95% CI shaded)"
        size = 800 to 400
    }
}
```

```kotlin
val xBetaEx = (0..100).map { it / 100.0 }
val beta22 = BetaDistribution(alpha = 2.0, beta = 5.0)
val beta52 = BetaDistribution(alpha = 5.0, beta = 2.0)
val beta55 = BetaDistribution(alpha = 5.0, beta = 5.0)
val beta11 = BetaDistribution(alpha = 1.0, beta = 1.0)

val allXB = xBetaEx + xBetaEx + xBetaEx + xBetaEx
val allYB = xBetaEx.map { beta22.pdf(it) } + xBetaEx.map { beta52.pdf(it) } + xBetaEx.map { beta55.pdf(it) } + xBetaEx.map { beta11.pdf(it) }
val allLabel = xBetaEx.map { "alpha=2, beta=5 (left-skewed)" } + xBetaEx.map { "alpha=5, beta=2 (right-skewed)" } + xBetaEx.map { "alpha=5, beta=5 (symmetric)" } + xBetaEx.map { "alpha=1, beta=1 (uniform)" }

plot {
    line {
        x(allXB); y(allYB)
        color(allLabel)
        width = 2.0
    }
    layout {
        title = "Beta Distribution — Shape Exploration"
        size = 800 to 400
    }
}
```

## General-Purpose Symmetric

```kotlin
val sessionDuration = NormalDistribution(mu = 12.5, sigma = 3.2)

println("P(session < 15 min): ${sessionDuration.cdf(15.0)}")
println("95th percentile: ${sessionDuration.quantile(0.95)} min")

val smallSampleEstimate = StudentTDistribution(degreesOfFreedom = 8.0)
println("Critical value for 95% CI: ${smallSampleEstimate.quantile(0.975)}")
```

```kotlin
val xNorm = (-50..50).map { it / 10.0 } // -5 to 5
val normal01 = NormalDistribution(mu = 0.0, sigma = 1.0)
val t3 = StudentTDistribution(degreesOfFreedom = 3.0)
val t8 = StudentTDistribution(degreesOfFreedom = 8.0)
val t30 = StudentTDistribution(degreesOfFreedom = 30.0)

val allXN = xNorm + xNorm + xNorm + xNorm
val allYN = xNorm.map { normal01.pdf(it) } + xNorm.map { t3.pdf(it) } + xNorm.map { t8.pdf(it) } + xNorm.map { t30.pdf(it) }
val allLabelN = xNorm.map { "Normal(0,1)" } + xNorm.map { "Student-t(df=3)" } + xNorm.map { "Student-t(df=8)" } + xNorm.map { "Student-t(df=30)" }

plot {
    line {
        x(allXN); y(allYN)
        color(allLabelN)
        width = 2.0
    }
    layout {
        title = "Normal vs Student-t — Heavier Tails with Fewer Degrees of Freedom"
        size = 800 to 400
    }
}
```

## Verifying the Fit

```kotlin
val processingTimesMs = doubleArrayOf(
    45.2, 51.8, 48.1, 52.3, 47.6, 49.9, 53.1, 46.5, 50.7, 48.8,
    51.2, 47.3, 49.1, 52.8, 46.9, 50.3, 48.5, 51.6, 47.8, 49.4
)

val fitted = NormalDistribution(
    mu = processingTimesMs.mean(),
    sigma = processingTimesMs.standardDeviation()
)

val ks = kolmogorovSmirnovTest(processingTimesMs, fitted)
println("KS statistic: ${ks.statistic}")
println("p-value: ${ks.pValue}")
println("Fit is plausible: ${ks.pValue > 0.05}")
```

```kotlin
val xFit = (400..550).map { it / 10.0 } // 40.0 to 55.0
val yFit = xFit.map { fitted.pdf(it) }

plot {
    densityPlot(processingTimesMs.toList()) {
        fillColor = Color.hex("#a5d8ff")
        alpha = 0.5
    }
    line {
        x(xFit); y(yFit)
        color = Color.hex("#e03131")
        width = 2.0
    }
    layout {
        title = "Kernel Density vs Fitted Normal (KS p=${String.format("%.3f", ks.pValue)})"
        size = 800 to 400
    }
}
```

```kotlin
val sortedData = processingTimesMs.sorted()
val n = sortedData.size
val empiricalX = sortedData.toList()
val empiricalY = (1..n).map { it.toDouble() / n }
val theoreticalY = empiricalX.map { fitted.cdf(it) }

val allXCdf = empiricalX + empiricalX
val allYCdf = empiricalY + theoreticalY
val allLabelCdf = empiricalX.map { "Empirical CDF" } + empiricalX.map { "Theoretical CDF (Normal)" }

plot {
    line {
        x(allXCdf); y(allYCdf)
        color(allLabelCdf)
        width = 2.0
    }
    points {
        x(empiricalX); y(empiricalY)
        color = Color.hex("#1971c2")
        size = 3.0
    }
    layout {
        title = "Empirical vs Theoretical CDF"
        size = 800 to 400
    }
}
```

## Summary

> A non-significant KS test does not prove the distribution is correct — it means the data does not strongly contradict that choice.
