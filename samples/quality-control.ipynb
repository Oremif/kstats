# Quality Control with kstats

Detect anomalies, set control limits, and monitor process stability using statistical methods.
We use temperature sensor readings from a manufacturing line.

## Process Data

```kotlin
val sensorReadings = doubleArrayOf(
    155.2, 154.8, 156.1, 155.5, 154.3, 155.9, 155.0, 156.3, 154.7, 155.4,
    155.8, 154.5, 156.0, 155.3, 154.9, 155.7, 155.1, 156.2, 154.6, 155.6,
    155.3, 154.4, 155.8, 162.5, 155.1, 155.7, 154.2, 155.5, 155.9, 141.3
)
// Two potential outliers: reading #24 (162.5°C) and #30 (141.3°C)
val readingIndex = (1..sensorReadings.size).toList()
```

```kotlin
val normal = sensorReadings.indices.filter { sensorReadings[it] in 150.0..160.0 }
val outlierIdx = sensorReadings.indices.filter { sensorReadings[it] !in 150.0..160.0 }

plot {
    line {
        x(readingIndex.map { it.toDouble() }); y(sensorReadings.toList())
        color = Color.hex("#1971c2")
        width = 1.5
    }
    points {
        x(normal.map { it + 1.0 }); y(normal.map { sensorReadings[it] })
        color = Color.hex("#1971c2")
        size = 4.0
    }
    points {
        x(outlierIdx.map { it + 1.0 }); y(outlierIdx.map { sensorReadings[it] })
        color = Color.hex("#e03131")
        size = 7.0
    }
    layout { title = "Sensor Readings \u2014 Red Points = Potential Outliers"; size = 800 to 400 }
}
```

## Anomaly Detection

### Baseline Statistics

```kotlin
val summary = sensorReadings.describe()
println("Mean: %.2f".format(summary.mean))
println("SD: %.2f".format(summary.standardDeviation))
println("Min: %.1f, Max: %.1f".format(summary.min, summary.max))
println("IQR: %.2f".format(summary.interquartileRange))
```

### Z-Score Method

```kotlin
val zScores = sensorReadings.zScore()
val anomalyIndices = zScores.indices.filter { kotlin.math.abs(zScores[it]) > 3.0 }
val anomalies = anomalyIndices.map { sensorReadings[it] }
println("Anomalies (|z| > 3): indices=$anomalyIndices, values=$anomalies")
```

```kotlin
plot {
    bars {
        x(readingIndex); y(zScores.toList())
        fillColor = Color.hex("#4dabf7")
        alpha = 0.7
    }
    line {
        x(listOf(1.0, 30.0)); y(listOf(3.0, 3.0))
        color = Color.hex("#e03131")
        width = 1.5
    }
    line {
        x(listOf(1.0, 30.0)); y(listOf(-3.0, -3.0))
        color = Color.hex("#e03131")
        width = 1.5
    }
    layout { title = "Z-Scores \u2014 Red Lines at \u00b13\u03c3"; size = 800 to 400 }
}
```

### Percentile-Based Detection

```kotlin
val lowerBound = sensorReadings.quantile(0.01)
val upperBound = sensorReadings.quantile(0.99)
val outliers = sensorReadings.filter { it < lowerBound || it > upperBound }
println("Bounds: [${"%1.1f".format(lowerBound)}, ${"%1.1f".format(upperBound)}]")
println("Outliers: $outliers")
```

```kotlin
plot {
    // Lower bound line
    line {
        x(listOf(1.0, 30.0)); y(listOf(lowerBound, lowerBound))
        color = Color.hex("#69db7c")
        width = 1.5
    }
    // Upper bound line
    line {
        x(listOf(1.0, 30.0)); y(listOf(upperBound, upperBound))
        color = Color.hex("#69db7c")
        width = 1.5
    }
    line {
        x(readingIndex.map { it.toDouble() }); y(sensorReadings.toList())
        color = Color.hex("#1971c2")
        width = 1.5
    }
    points {
        x(readingIndex.map { it.toDouble() }); y(sensorReadings.toList())
        color = Color.hex("#1971c2")
        size = 3.0
    }
    layout { title = "Sensor Readings with 1st\u201399th Percentile Band"; size = 800 to 400 }
}
```

## Control Limits

```kotlin
val centerLine = sensorReadings.mean()
val sigma = sensorReadings.standardDeviation()
val upperControlLimit = centerLine + 3 * sigma
val lowerControlLimit = centerLine - 3 * sigma
val outOfControl = sensorReadings.filter { it > upperControlLimit || it < lowerControlLimit }

println("Center: %.2f".format(centerLine))
println("UCL: %.2f, LCL: %.2f".format(upperControlLimit, lowerControlLimit))
println("Out of control: $outOfControl")
```

```kotlin
val oocIdx = sensorReadings.indices.filter { sensorReadings[it] > upperControlLimit || sensorReadings[it] < lowerControlLimit }
val inCtrl = sensorReadings.indices.filter { sensorReadings[it] in lowerControlLimit..upperControlLimit }

plot {
    line {
        x(readingIndex.map { it.toDouble() }); y(sensorReadings.toList())
        color = Color.hex("#495057")
        width = 1.5
    }
    points {
        x(inCtrl.map { it + 1.0 }); y(inCtrl.map { sensorReadings[it] })
        color = Color.hex("#1971c2")
        size = 4.0
    }
    points {
        x(oocIdx.map { it + 1.0 }); y(oocIdx.map { sensorReadings[it] })
        color = Color.hex("#e03131")
        size = 7.0
    }
    // Center line
    line {
        x(listOf(1.0, 30.0)); y(listOf(centerLine, centerLine))
        color = Color.hex("#2b8a3e")
        width = 1.5
    }
    // UCL
    line {
        x(listOf(1.0, 30.0)); y(listOf(upperControlLimit, upperControlLimit))
        color = Color.hex("#e03131")
        width = 1.5
    }
    // LCL
    line {
        x(listOf(1.0, 30.0)); y(listOf(lowerControlLimit, lowerControlLimit))
        color = Color.hex("#e03131")
        width = 1.5
    }
    layout { title = "Control Chart \u2014 Green=Center, Red=UCL/LCL"; size = 900 to 450 }
}
```

## Batch Stability

```kotlin
val morningBatch = doubleArrayOf(
    155.2, 154.8, 156.1, 155.5, 154.3, 155.9, 155.0, 156.3, 154.7, 155.4
)
val eveningBatch = doubleArrayOf(
    155.8, 154.5, 156.0, 155.3, 154.9, 155.7, 155.1, 156.2, 154.6, 155.6
)

println("Morning normality p=${String.format("%.4f", shapiroWilkTest(morningBatch).pValue)}")
println("Evening normality p=${String.format("%.4f", shapiroWilkTest(eveningBatch).pValue)}")
println("Levene p=${String.format("%.4f", leveneTest(morningBatch, eveningBatch).pValue)}")

val batchComparison = tTest(morningBatch, eveningBatch)
println("t-test p=${String.format("%.4f", batchComparison.pValue)}, significant=${batchComparison.isSignificant()}")
```

```kotlin
plot {
    densityPlot(morningBatch.toList()) {
        fillColor = Color.hex("#4dabf7")
        alpha = 0.4
    }
    densityPlot(eveningBatch.toList()) {
        fillColor = Color.hex("#ff8787")
        alpha = 0.4
    }
    layout { title = "Morning (blue) vs Evening (red) Batch Density"; size = 800 to 400 }
}
```

### Multiple Batches (ANOVA)

```kotlin
val batch1 = doubleArrayOf(155.2, 154.8, 156.1, 155.5, 154.3)
val batch2 = doubleArrayOf(155.9, 155.0, 156.3, 154.7, 155.4)
val batch3 = doubleArrayOf(155.8, 154.5, 156.0, 155.3, 154.9)

val anova = oneWayAnova(batch1, batch2, batch3)
println("F-statistic: ${String.format("%.3f", anova.fStatistic)}")
println("p-value: ${String.format("%.4f", anova.pValue)}")
```

```kotlin
val allBatch3 = batch1.toList() + batch2.toList() + batch3.toList()
val allBatch3Label = batch1.map { "Batch 1" } + batch2.map { "Batch 2" } + batch3.map { "Batch 3" }

plot {
    points {
        x(allBatch3Label); y(allBatch3)
        color(allBatch3Label)
        size = 5.0
    }
    layout { title = "Three Batches \u2014 Individual Readings (ANOVA p=${String.format("%.3f", anova.pValue)})"; size = 700 to 400 }
}
```

## Distribution Fit

```kotlin
val cleanReadings = sensorReadings.filter { it in 150.0..160.0 }.toDoubleArray()

val fitted = NormalDistribution(
    mu = cleanReadings.mean(),
    sigma = cleanReadings.standardDeviation()
)

val ks = kolmogorovSmirnovTest(cleanReadings, fitted)
println("KS statistic: ${String.format("%.4f", ks.statistic)}")
println("p-value: ${String.format("%.4f", ks.pValue)}")
println("Process threshold (0.1%): ${String.format("%.1f", fitted.quantile(0.001))}")
println("Process threshold (99.9%): ${String.format("%.1f", fitted.quantile(0.999))}")
```

```kotlin
val xFit = (1530..1570).map { it / 10.0 }
val yFitPdf = xFit.map { fitted.pdf(it) }
val lo = fitted.quantile(0.001)
val hi = fitted.quantile(0.999)

plot {
    densityPlot(cleanReadings.toList()) {
        fillColor = Color.hex("#a5d8ff")
        alpha = 0.5
    }
    line {
        x(xFit); y(yFitPdf)
        color = Color.hex("#e03131")
        width = 2.0
    }
    line {
        x(listOf(lo, lo)); y(listOf(0.0, yFitPdf.max()!!))
        color = Color.hex("#f08c00")
        width = 1.5
    }
    line {
        x(listOf(hi, hi)); y(listOf(0.0, yFitPdf.max()!!))
        color = Color.hex("#f08c00")
        width = 1.5
    }
    layout { title = "Fitted Normal + 99.8% Coverage Thresholds (orange)"; size = 800 to 400 }
}
```

## Multi-Parameter Monitoring

```kotlin
val temperatureSensor = doubleArrayOf(
    155.2, 154.8, 156.1, 155.5, 154.3, 155.9, 155.0, 156.3, 154.7, 155.4
)
val pressureSensor = doubleArrayOf(
    2.12, 2.08, 2.15, 2.11, 2.05, 2.14, 2.09, 2.16, 2.07, 2.10
)
val flowRateSensor = doubleArrayOf(
    45.1, 44.8, 45.5, 45.2, 44.5, 45.4, 44.9, 45.6, 44.7, 45.3
)

println("Temperature: ${temperatureSensor.describe().let { "mean=${String.format("%.2f", it.mean)}, sd=${String.format("%.2f", it.standardDeviation)}" }}")
println("Pressure:    ${pressureSensor.describe().let { "mean=${String.format("%.3f", it.mean)}, sd=${String.format("%.3f", it.standardDeviation)}" }}")
println("Flow Rate:   ${flowRateSensor.describe().let { "mean=${String.format("%.2f", it.mean)}, sd=${String.format("%.2f", it.standardDeviation)}" }}")

val sensorMatrix = correlationMatrix(temperatureSensor, pressureSensor, flowRateSensor)
val sLabels = listOf("Temp", "Pressure", "Flow")
for (i in sLabels.indices) {
    println("${sLabels[i]}: ${sensorMatrix[i].map { String.format("%.3f", it) }}")
}
```

```kotlin
val sHeatX = mutableListOf<String>()
val sHeatY = mutableListOf<String>()
val sHeatV = mutableListOf<Double>()
for (i in sLabels.indices) {
    for (j in sLabels.indices) {
        sHeatX.add(sLabels[j])
        sHeatY.add(sLabels[i])
        sHeatV.add(sensorMatrix[i][j])
    }
}

plot {
    tiles {
        x(sHeatX); y(sHeatY)
        fillColor(sHeatV)
    }
    layout { title = "Sensor Correlation Matrix"; size = 500 to 450 }
}
```
