# Testing Assumptions

Parametric methods (t-tests, ANOVA, Pearson correlation) assume specific properties of the data.
This notebook covers **normality**, **variance homogeneity**, and **goodness-of-fit**.

## Normality

| Test | Function | Best for |
| --- | --- | --- |
| Shapiro-Wilk | `shapiroWilkTest(sample)` | General-purpose default, up to ~5000 |
| Anderson-Darling | `andersonDarlingTest(sample)` | Sensitive to tail deviations |
| D'Agostino-Pearson | `dagostinoPearsonTest(sample)` | Omnibus (skewness+kurtosis), n>20 |
| Jarque-Bera | `jarqueBeraTest(sample)` | Large samples |

### Run All Four Tests

```kotlin
val sensorReadings = doubleArrayOf(
    150.2, 151.8, 149.6, 152.1, 150.9, 151.3, 149.8, 152.5, 150.4, 151.1,
    150.7, 149.5, 151.6, 150.0, 152.3, 151.0, 149.9, 150.8, 151.5, 150.3,
    151.2, 149.7, 150.6, 152.0, 150.1, 151.4, 149.4, 151.9, 150.5, 151.7
)

val shapiro = shapiroWilkTest(sensorReadings)
val anderson = andersonDarlingTest(sensorReadings)
val dagostino = dagostinoPearsonTest(sensorReadings)
val jarqueBera = jarqueBeraTest(sensorReadings)

println("Shapiro-Wilk:      p=${String.format("%.4f", shapiro.pValue)}")
println("Anderson-Darling:  p=${String.format("%.4f", anderson.pValue)}")
println("D'Agostino-Pearson: p=${String.format("%.4f", dagostino.pValue)}")
println("Jarque-Bera:       p=${String.format("%.4f", jarqueBera.pValue)}")
```

```kotlin
val testNames = listOf("Shapiro-Wilk", "Anderson-Darling", "D'Agostino-Pearson", "Jarque-Bera")
val pValues = listOf(shapiro.pValue, anderson.pValue, dagostino.pValue, jarqueBera.pValue)

plot {
    bars {
        x(testNames); y(pValues)
        fillColor = Color.hex("#4dabf7")
        alpha = 0.8
    }
    line {
        x(listOf(testNames.first(), testNames.last())); y(listOf(0.05, 0.05))
        color = Color.hex("#e03131")
        width = 2.0
    }
    layout { title = "Normality Test p-values (red = \u03b1=0.05)"; size = 800 to 400 }
}
```

```kotlin
val mu = sensorReadings.mean()
val sigma = sensorReadings.standardDeviation()
val fittedNormal = NormalDistribution(mu = mu, sigma = sigma)

val xFit = (1490..1530).map { it / 10.0 }
val yFit = xFit.map { fittedNormal.pdf(it) }

plot {
    densityPlot(sensorReadings.toList()) {
        fillColor = Color.hex("#a5d8ff")
        alpha = 0.5
    }
    line {
        x(xFit); y(yFit)
        color = Color.hex("#e03131")
        width = 2.0
    }
    layout { title = "Sensor Data: Kernel Density vs Fitted Normal"; size = 800 to 400 }
}
```

### Descriptive Shape Indicators

> When tests disagree, prefer Shapiro-Wilk for small to medium samples and Anderson-Darling when tail behavior matters.

```kotlin
val summary = sensorReadings.describe()
println("Skewness: ${String.format("%.4f", summary.skewness)} (close to 0 = symmetric)")
println("Kurtosis: ${String.format("%.4f", summary.kurtosis)} (close to 0 = Normal-like tails)")
```

```kotlin
val normalDist = NormalDistribution(mu = 0.0, sigma = 1.0)
val skewedDist = LogNormalDistribution(mu = 0.0, sigma = 0.8)
val heavyDist = StudentTDistribution(degreesOfFreedom = 3.0)

val xRef = (-40..60).map { it / 10.0 }
val allXRef = xRef + (0..100).map { it / 10.0 } + xRef
val allYRef = xRef.map { normalDist.pdf(it) } + (0..100).map { it / 10.0 }.map { skewedDist.pdf(it) } + xRef.map { heavyDist.pdf(it) }
val allLabelRef = xRef.map { "Normal (symmetric)" } + (0..100).map { "LogNormal (right-skewed)" } + xRef.map { "Student-t df=3 (heavy tails)" }

plot {
    line {
        x(allXRef); y(allYRef)
        color(allLabelRef)
        width = 2.0
    }
    layout { title = "Shape Reference: Normal vs Skewed vs Heavy-Tailed"; size = 800 to 400 }
}
```

## Variance Homogeneity

| Test | Function | Assumes normality? |
| --- | --- | --- |
| Levene | `leveneTest(...)` | No — robust |
| Bartlett | `bartlettTest(...)` | Yes — powerful when normal |
| Fligner-Killeen | `flignerKilleenTest(...)` | No — non-parametric |

### Check Variances Before ANOVA

```kotlin
val batchA = doubleArrayOf(48.2, 47.8, 49.1, 48.5, 47.9, 48.7, 48.3, 49.0, 48.1, 48.6)
val batchB = doubleArrayOf(51.3, 50.8, 52.1, 51.0, 51.7, 50.5, 51.9, 51.2, 50.9, 51.5)
val batchC = doubleArrayOf(49.5, 50.2, 49.8, 50.0, 49.3, 50.4, 49.7, 50.1, 49.6, 50.3)

val levene = leveneTest(batchA, batchB, batchC)
val bartlett = bartlettTest(batchA, batchB, batchC)
val fligner = flignerKilleenTest(batchA, batchB, batchC)

println("Levene:          p=${String.format("%.4f", levene.pValue)}")
println("Bartlett:        p=${String.format("%.4f", bartlett.pValue)}")
println("Fligner-Killeen: p=${String.format("%.4f", fligner.pValue)}")
```

```kotlin
plot {
    bars {
        x(listOf("Levene", "Bartlett", "Fligner-Killeen")); y(listOf(levene.pValue, bartlett.pValue, fligner.pValue))
        fillColor = Color.hex("#69db7c")
        alpha = 0.8
    }
    line {
        x(listOf("Levene", "Fligner-Killeen")); y(listOf(0.05, 0.05))
        color = Color.hex("#e03131")
        width = 2.0
    }
    layout { title = "Variance Homogeneity Test p-values (red = \u03b1=0.05)"; size = 700 to 400 }
}
```

```kotlin
plot {
    densityPlot(batchA.toList()) {
        fillColor = Color.hex("#4dabf7")
        alpha = 0.3
    }
    densityPlot(batchB.toList()) {
        fillColor = Color.hex("#ff8787")
        alpha = 0.3
    }
    densityPlot(batchC.toList()) {
        fillColor = Color.hex("#69db7c")
        alpha = 0.3
    }
    layout { title = "Batch A (blue) / B (red) / C (green) — Visual Spread Comparison"; size = 800 to 400 }
}
```

```kotlin
val allBatchData = batchA.toList() + batchB.toList() + batchC.toList()
val allBatchLabels = batchA.map { "Batch A" } + batchB.map { "Batch B" } + batchC.map { "Batch C" }

plot {
    points {
        x(allBatchLabels); y(allBatchData)
        color(allBatchLabels)
        size = 5.0
    }
    layout { title = "Batch Data: Individual Observations"; size = 600 to 400 }
}
```

### Then Run ANOVA

```kotlin
val anova = oneWayAnova(batchA, batchB, batchC)
println("F-statistic: ${String.format("%.2f", anova.fStatistic)}")
println("p-value: ${String.format("%.6f", anova.pValue)}")
```

```kotlin
plot {
    bars {
        x(listOf("Batch A", "Batch B", "Batch C")); y(listOf(batchA.mean(), batchB.mean(), batchC.mean()))
        fillColor = Color.hex("#9775fa")
        alpha = 0.8
    }
    layout { title = "Batch Means (ANOVA F=${String.format("%.2f", anova.fStatistic)}, p=${String.format("%.4f", anova.pValue)})"; size = 600 to 400 }
}
```

## Goodness-of-Fit

### Kolmogorov-Smirnov Test

```kotlin
val temperatureReadings = doubleArrayOf(
    155.2, 154.8, 156.1, 155.5, 154.3, 155.9, 155.0, 156.3, 154.7, 155.4,
    155.8, 154.5, 156.0, 155.3, 154.9, 155.7, 155.1, 156.2, 154.6, 155.6
)

val fitted = NormalDistribution(
    mu = temperatureReadings.mean(),
    sigma = temperatureReadings.standardDeviation()
)

val ks = kolmogorovSmirnovTest(temperatureReadings, fitted)
println("KS statistic: ${String.format("%.4f", ks.statistic)}")
println("p-value: ${String.format("%.4f", ks.pValue)}")
```

```kotlin
val sortedTemp = temperatureReadings.sorted()
val n = sortedTemp.size
val empiricalY = (1..n).map { it.toDouble() / n }
val theoreticalY = sortedTemp.map { fitted.cdf(it) }

val allXCdf = sortedTemp + sortedTemp
val allYCdf = empiricalY + theoreticalY
val allLabelCdf = sortedTemp.map { "Empirical CDF" } + sortedTemp.map { "Theoretical CDF (Normal)" }

plot {
    line {
        x(allXCdf); y(allYCdf)
        color(allLabelCdf)
        width = 2.0
    }
    points {
        x(sortedTemp.toList()); y(empiricalY)
        color = Color.hex("#1971c2")
        size = 3.0
    }
    layout { title = "Empirical vs Theoretical CDF (KS=${String.format("%.4f", ks.statistic)})"; size = 800 to 400 }
}
```

### Chi-Squared Goodness-of-Fit

```kotlin
val observedDefects = intArrayOf(12, 18, 25, 15, 30)

val uniform = chiSquaredTest(observedDefects)
println("Uniform test p=${String.format("%.4f", uniform.pValue)}")

val expectedCounts = doubleArrayOf(20.0, 20.0, 20.0, 20.0, 20.0)
val specific = chiSquaredTest(observedDefects, expectedCounts)
println("Specific expected p=${String.format("%.4f", specific.pValue)}")
```

```kotlin
val categories = listOf("Cat A", "Cat B", "Cat C", "Cat D", "Cat E")
val allCatX = categories + categories
val allCatY = observedDefects.map { it.toDouble() } + expectedCounts.toList()
val allCatLabel = categories.map { "Observed" } + categories.map { "Expected" }

plot {
    bars {
        x(allCatX); y(allCatY)
        fillColor(allCatLabel)
        alpha = 0.8
    }
    layout { title = "Observed vs Expected Defect Counts (\u03c7\u00b2 p=${String.format("%.4f", specific.pValue)})"; size = 800 to 400 }
}
```

### Two-Sample KS Test

```kotlin
val morningReadings = doubleArrayOf(
    155.2, 154.8, 156.1, 155.5, 154.3, 155.9, 155.0, 156.3, 154.7, 155.4
)
val nightReadings = doubleArrayOf(
    156.1, 155.3, 157.0, 156.5, 155.8, 156.8, 155.5, 157.2, 155.9, 156.3
)

val twoSampleKs = kolmogorovSmirnovTest(morningReadings, nightReadings)
println("Two-sample KS p=${String.format("%.4f", twoSampleKs.pValue)}")
```

```kotlin
val sortedMorning = morningReadings.sorted()
val sortedNight = nightReadings.sorted()
val empMorning = (1..sortedMorning.size).map { it.toDouble() / sortedMorning.size }
val empNight = (1..sortedNight.size).map { it.toDouble() / sortedNight.size }

val allX2 = sortedMorning + sortedNight
val allY2 = empMorning + empNight
val allLabel2 = sortedMorning.map { "Morning" } + sortedNight.map { "Night" }

plot {
    line {
        x(allX2); y(allY2)
        color(allLabel2)
        width = 2.0
    }
    points {
        x(allX2); y(allY2)
        color(allLabel2)
        size = 3.0
    }
    layout { title = "Morning vs Night: Empirical CDFs"; size = 800 to 400 }
}
```

```kotlin
plot {
    densityPlot(morningReadings.toList()) {
        fillColor = Color.hex("#4dabf7")
        alpha = 0.4
    }
    densityPlot(nightReadings.toList()) {
        fillColor = Color.hex("#ff8787")
        alpha = 0.4
    }
    layout { title = "Morning (blue) vs Night (red): Density Comparison"; size = 800 to 400 }
}
```

## Summary

Always test assumptions before applying parametric methods. When normality or equal variances are violated, switch to non-parametric alternatives.
