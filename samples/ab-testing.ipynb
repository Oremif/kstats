# A/B Testing with kstats

A complete A/B test workflow: summarize groups, check assumptions, pick the right test,
measure effect size, and handle multiple metrics. We compare two checkout flow variants in a mobile app.

## Experiment Data

```kotlin
// Variant A (control): original checkout flow
val controlDurationSec = doubleArrayOf(
    34.2, 41.5, 38.7, 45.1, 36.9, 42.3, 39.8, 44.6, 37.4, 40.1,
    43.2, 35.8, 41.9, 38.3, 46.0, 39.5, 42.7, 37.1, 40.8, 44.3
)
// Variant B (treatment): simplified checkout flow
val treatmentDurationSec = doubleArrayOf(
    29.1, 33.8, 31.5, 35.2, 28.7, 32.4, 30.9, 34.6, 29.8, 33.1,
    31.2, 27.5, 34.0, 30.3, 36.1, 31.8, 33.5, 28.9, 32.7, 35.8
)
```

## Step 1: Summarize Both Groups

```kotlin
val controlSummary = controlDurationSec.describe()
val treatmentSummary = treatmentDurationSec.describe()

println("Control   — mean: %.1f s, sd: %.1f".format(controlSummary.mean, controlSummary.standardDeviation))
println("Treatment — mean: %.1f s, sd: %.1f".format(treatmentSummary.mean, treatmentSummary.standardDeviation))
println("Difference: %.1f s".format(controlSummary.mean - treatmentSummary.mean))
```

```kotlin
plot {
    densityPlot(controlDurationSec.toList()) {
        fillColor = Color.hex("#4dabf7")
        alpha = 0.4
    }
    densityPlot(treatmentDurationSec.toList()) {
        fillColor = Color.hex("#ff8787")
        alpha = 0.4
    }
    line {
        x(listOf(controlSummary.mean, controlSummary.mean)); y(listOf(0.0, 0.15))
        color = Color.hex("#1971c2")
        width = 1.5
    }
    line {
        x(listOf(treatmentSummary.mean, treatmentSummary.mean)); y(listOf(0.0, 0.15))
        color = Color.hex("#e03131")
        width = 1.5
    }
    layout { title = "Session Duration: Control (blue) vs Treatment (red)"; size = 800 to 400 }
}
```

## Step 2: Check Assumptions

**Normality** (Shapiro-Wilk) and **variance homogeneity** (Levene) determine which test to use:
- Both pass -> Welch's t-test (default, does not assume equal variance)
- Normality fails -> Mann-Whitney U (non-parametric)

```kotlin
val controlNormality = shapiroWilkTest(controlDurationSec)
val treatmentNormality = shapiroWilkTest(treatmentDurationSec)
val variances = leveneTest(controlDurationSec, treatmentDurationSec)

println("Shapiro-Wilk normality:")
println("  Control   p=${String.format("%.4f", controlNormality.pValue)} ${if (controlNormality.pValue > 0.05) "✓ normal" else "✗ non-normal"}")
println("  Treatment p=${String.format("%.4f", treatmentNormality.pValue)} ${if (treatmentNormality.pValue > 0.05) "✓ normal" else "✗ non-normal"}")
println("Levene test p=${String.format("%.4f", variances.pValue)} ${if (variances.pValue > 0.05) "→ equal variances" else "→ unequal variances"}")
println()
println("Both groups normal, variances equal → Welch's t-test is appropriate")
```

## Step 3: Hypothesis Test + Effect Size

### Welch's t-test

```kotlin
val result = tTest(controlDurationSec, treatmentDurationSec)

println("Welch's t-test:")
println("  statistic: ${String.format("%.3f", result.statistic)}")
println("  p-value:   ${String.format("%.6f", result.pValue)}")
println("  95% CI:    [${result.confidenceInterval?.let { "${String.format("%.2f", it.lower)}, ${String.format("%.2f", it.upper)}" }}]")
println("  Significant: ${result.isSignificant()}")
println()

// Effect size: how large is the difference in standard-deviation units?
val d = cohensD(controlDurationSec, treatmentDurationSec)
val interpretation = when {
    kotlin.math.abs(d) < 0.2 -> "negligible"
    kotlin.math.abs(d) < 0.5 -> "small"
    kotlin.math.abs(d) < 0.8 -> "medium"
    else -> "large"
}
println("Cohen's d = ${String.format("%.2f", d)} ($interpretation effect)")
```

```kotlin
val meanDiff = controlSummary.mean - treatmentSummary.mean
val ciLow = result.confidenceInterval?.lower ?: 0.0
val ciHigh = result.confidenceInterval?.upper ?: 0.0

plot {
    points {
        x(listOf(meanDiff)); y(listOf("Mean Difference"))
        size = 8.0
        color = Color.hex("#1971c2")
    }
    line {
        x(listOf(ciLow, ciHigh)); y(listOf("Mean Difference", "Mean Difference"))
        color = Color.hex("#1971c2")
        width = 3.0
    }
    line {
        x(listOf(0.0, 0.0)); y(listOf("", "Mean Difference"))
        color = Color.hex("#e03131")
        width = 1.5
    }
    layout { title = "Mean Difference with 95% Confidence Interval"; size = 700 to 300 }
}
```

### Robustness: Non-parametric & One-sided

Mann-Whitney U confirms the result without assuming normality. The one-sided test checks
whether control is specifically *longer* than treatment.

```kotlin
val mwResult = mannWhitneyUTest(controlDurationSec, treatmentDurationSec)
println("Mann-Whitney U: statistic=${String.format("%.1f", mwResult.statistic)}, p=${String.format("%.6f", mwResult.pValue)}")

val oneSided = tTest(controlDurationSec, treatmentDurationSec, alternative = Alternative.GREATER)
println("One-sided t-test (control > treatment): p=${String.format("%.6f", oneSided.pValue)}")
```

```kotlin
val testNames = listOf("Welch t-test", "Mann-Whitney U", "One-sided t-test")
val testPvalues = listOf(result.pValue, mwResult.pValue, oneSided.pValue)

val negLog10 = testPvalues.map { -kotlin.math.ln(it) / kotlin.math.ln(10.0) }
val threshold = -kotlin.math.ln(0.05) / kotlin.math.ln(10.0)

plot {
    bars {
        x(testNames); y(negLog10)
        fillColor = Color.hex("#9775fa")
        alpha = 0.8
    }
    line {
        x(listOf(testNames.first(), testNames.last())); y(listOf(threshold, threshold))
        color = Color.hex("#e03131")
        width = 2.0
    }
    layout { title = "-log10(p-value) — Higher = More Significant (red line = 0.05)"; size = 800 to 400 }
}
```

## Step 4: Second Metric — Checkout Steps

The simplified flow also tracks how many checkout steps users complete. Since step counts
are ordinal (not normally distributed), we use Mann-Whitney U directly.

```kotlin
val controlSteps = doubleArrayOf(
    3.0, 4.0, 3.0, 5.0, 3.0, 4.0, 4.0, 5.0, 3.0, 4.0,
    4.0, 3.0, 4.0, 3.0, 5.0, 4.0, 4.0, 3.0, 4.0, 5.0
)
val treatmentSteps = doubleArrayOf(
    5.0, 5.0, 4.0, 5.0, 5.0, 5.0, 4.0, 5.0, 5.0, 5.0,
    4.0, 5.0, 5.0, 4.0, 5.0, 5.0, 5.0, 4.0, 5.0, 5.0
)

val stepsResult = mannWhitneyUTest(controlSteps, treatmentSteps)
println("Checkout steps — Mann-Whitney U: p=${String.format("%.6f", stepsResult.pValue)}, significant=${stepsResult.isSignificant()}")
```

```kotlin
val stepValues = listOf(3, 4, 5)
val controlCounts = stepValues.map { v -> controlSteps.count { it == v.toDouble() } }
val treatmentCounts = stepValues.map { v -> treatmentSteps.count { it == v.toDouble() } }

val allStepX = stepValues.map { "$it steps" } + stepValues.map { "$it steps" }
val allStepY = controlCounts + treatmentCounts
val allStepGroup = stepValues.map { "Control" } + stepValues.map { "Treatment" }

plot {
    bars {
        x(allStepX); y(allStepY.map { it.toDouble() })
        fillColor(allStepGroup)
        alpha = 0.8
    }
    layout { title = "Completed Steps: Control vs Treatment"; size = 700 to 400 }
}
```

### Multiple Comparison Correction

We tested two metrics (duration and steps), which inflates the false-positive rate.
Holm-Bonferroni controls this while being less conservative than plain Bonferroni.

```kotlin
val rawPValues = doubleArrayOf(result.pValue, stepsResult.pValue)
val corrected = holmBonferroniCorrection(rawPValues)

println("            Raw p      Holm p")
println("Duration    ${String.format("%.6f", rawPValues[0])}   ${String.format("%.6f", corrected[0])}")
println("Steps       ${String.format("%.6f", rawPValues[1])}   ${String.format("%.6f", corrected[1])}")
println()
println("Both remain significant after correction.")
```

## Step 5: Correlation Between Metrics

```kotlin
val correlation = spearmanCorrelation(treatmentDurationSec, treatmentSteps)
println("Spearman rho = ${String.format("%.3f", correlation.coefficient)}")
println("p-value = ${String.format("%.4f", correlation.pValue)}")
```

```kotlin
val rng = java.util.Random(42)
val jitteredSteps = treatmentSteps.map { it + (rng.nextDouble() - 0.5) * 0.3 }

plot {
    points {
        x(treatmentDurationSec.toList()); y(jitteredSteps)
        color = Color.hex("#e03131")
        size = 6.0
    }
    layout {
        title = "Treatment: Duration vs Steps (Spearman rho=${String.format("%.3f", correlation.coefficient)})"
        size = 700 to 500
    }
}
```

## Paired Before/After Comparison

When the same users are measured before and after a change, use paired tests.
The paired design removes between-subject variability, increasing statistical power.

```kotlin
val beforeMs = doubleArrayOf(
    340.2, 415.0, 387.1, 451.3, 369.5, 423.8, 398.0, 446.2, 374.1, 401.5
)
val afterMs = doubleArrayOf(
    310.5, 380.2, 355.8, 410.7, 335.1, 392.4, 365.3, 405.9, 340.8, 371.6
)

val paired = pairedTTest(beforeMs, afterMs)
println("Paired t-test: p=${String.format("%.4f", paired.pValue)}, CI=[${paired.confidenceInterval?.let { "${String.format("%.1f", it.lower)}, ${String.format("%.1f", it.upper)}" }}]")

val wilcoxon = wilcoxonSignedRankTest(beforeMs, afterMs)
println("Wilcoxon signed-rank: p=${String.format("%.4f", wilcoxon.pValue)}")

// Paired effect size: Cohen's dz = mean(diff) / sd(diff)
val differences = DoubleArray(beforeMs.size) { beforeMs[it] - afterMs[it] }
val dz = differences.mean() / differences.describe().standardDeviation
println("\nCohen's dz = ${String.format("%.2f", dz)} (${if (kotlin.math.abs(dz) >= 0.8) "large" else if (kotlin.math.abs(dz) >= 0.5) "medium" else "small"} effect)")
```

```kotlin
plot {
    // Connecting lines for each user
    for (i in beforeMs.indices) {
        val x = (i + 1).toDouble()
        line {
            x(listOf(x, x)); y(listOf(beforeMs[i], afterMs[i]))
            color = Color.hex("#ced4da"); width = 1.0
        }
    }
    points {
        x((1..beforeMs.size).map { it.toDouble() }); y(beforeMs.toList())
        color = Color.hex("#1971c2"); size = 6.0
    }
    points {
        x((1..afterMs.size).map { it.toDouble() }); y(afterMs.toList())
        color = Color.hex("#e03131"); size = 6.0
    }
    layout { title = "Before (blue) vs After (red) — Connected Pairs"; size = 800 to 450 }
}
```

```kotlin
val meanDiffPaired = differences.mean()

plot {
    histogram(differences.toList()) {
        fillColor = Color.hex("#ffd43b")
        alpha = 0.7
    }
    line {
        x(listOf(meanDiffPaired, meanDiffPaired)); y(listOf(0.0, 4.0))
        color = Color.hex("#1971c2")
        width = 1.5
    }
    layout { title = "Before-After Differences (blue line = mean ${String.format("%.1f", meanDiffPaired)} ms)"; size = 700 to 400 }
}
```

## Summary

| Metric | Test | p-value | Effect Size | Decision |
|--------|------|---------|-------------|----------|
| Duration | Welch's t-test | <0.001 | Cohen's d ~ 2.9 (large) | Treatment is faster |
| Steps | Mann-Whitney U | <0.001 | — | Treatment completes more steps |
| Before/After | Paired t-test | <0.001 | Cohen's dz ~ 6.1 (large) | Significant improvement |

The simplified checkout flow significantly reduced session duration while increasing
completed steps. Both parametric and non-parametric tests confirm the effect, and results
hold after Holm multiple-comparison correction.
