# A/B Testing with kstats

Run a complete A/B test workflow: summarize groups, check assumptions, choose the right test, and analyze multiple metrics. We compare two checkout flow variants in a mobile app.

## Experiment Data

```kotlin
// Variant A (control): original checkout flow
val controlDurationSec = doubleArrayOf(
    34.2, 41.5, 38.7, 45.1, 36.9, 42.3, 39.8, 44.6, 37.4, 40.1,
    43.2, 35.8, 41.9, 38.3, 46.0, 39.5, 42.7, 37.1, 40.8, 44.3
)
// Variant B (treatment): simplified checkout flow
val treatmentDurationSec = doubleArrayOf(
    29.1, 33.8, 31.5, 35.2, 28.7, 32.4, 30.9, 34.6, 29.8, 33.1,
    31.2, 27.5, 34.0, 30.3, 36.1, 31.8, 33.5, 28.9, 32.7, 35.8
)
```

## Step 1: Summarize Both Groups

```kotlin
val controlSummary = controlDurationSec.describe()
val treatmentSummary = treatmentDurationSec.describe()

println("Control   — mean: %.1f s, sd: %.1f".format(controlSummary.mean, controlSummary.standardDeviation))
println("Treatment — mean: %.1f s, sd: %.1f".format(treatmentSummary.mean, treatmentSummary.standardDeviation))
println("Difference: %.1f s".format(controlSummary.mean - treatmentSummary.mean))
```

```kotlin
plot {
    densityPlot(controlDurationSec.toList()) {
        fillColor = Color.hex("#4dabf7")
        alpha = 0.4
    }
    densityPlot(treatmentDurationSec.toList()) {
        fillColor = Color.hex("#ff8787")
        alpha = 0.4
    }
    line {
        x(listOf(controlSummary.mean, controlSummary.mean)); y(listOf(0.0, 0.15))
        color = Color.hex("#1971c2")
        width = 1.5
    }
    line {
        x(listOf(treatmentSummary.mean, treatmentSummary.mean)); y(listOf(0.0, 0.15))
        color = Color.hex("#e03131")
        width = 1.5
    }
    layout { title = "Session Duration: Control (blue) vs Treatment (red)"; size = 800 to 400 }
}
```

```kotlin
val allDurations = controlDurationSec.toList() + treatmentDurationSec.toList()
val allGroups = controlDurationSec.map { "Control" } + treatmentDurationSec.map { "Treatment" }

plot {
    points {
        x(allGroups); y(allDurations)
        color(allGroups)
        size = 5.0
    }
    layout { title = "Individual Session Durations by Group"; size = 600 to 400 }
}
```

## Step 2: Check Assumptions

### Normality

```kotlin
val controlNormality = shapiroWilkTest(controlDurationSec)
val treatmentNormality = shapiroWilkTest(treatmentDurationSec)

println("Control Shapiro-Wilk p=${String.format("%.4f", controlNormality.pValue)} ${if (controlNormality.pValue > 0.05) "✓" else "✗"}")
println("Treatment Shapiro-Wilk p=${String.format("%.4f", treatmentNormality.pValue)} ${if (treatmentNormality.pValue > 0.05) "✓" else "✗"}")
```

```kotlin
plot {
    histogram(controlDurationSec.toList()) {
        fillColor = Color.hex("#4dabf7")
        alpha = 0.6
    }
    layout { title = "Control Group Distribution"; size = 700 to 350 }
}
```

```kotlin
plot {
    histogram(treatmentDurationSec.toList()) {
        fillColor = Color.hex("#ff8787")
        alpha = 0.6
    }
    layout { title = "Treatment Group Distribution"; size = 700 to 350 }
}
```

### Variance Homogeneity

```kotlin
val variances = leveneTest(controlDurationSec, treatmentDurationSec)
println("Levene test p=${String.format("%.4f", variances.pValue)} ${if (variances.pValue > 0.05) "→ equal variances" else "→ unequal variances"}")
```

```kotlin
plot {
    bars {
        x(listOf("Control", "Treatment")); y(listOf(controlSummary.standardDeviation, treatmentSummary.standardDeviation))
        fillColor = Color.hex("#69db7c")
        alpha = 0.8
    }
    layout { title = "Standard Deviation Comparison"; size = 500 to 400 }
}
```

## Step 3: Choose and Run the Test

### Parametric: Welch's t-test

```kotlin
val result = tTest(controlDurationSec, treatmentDurationSec)

println("Welch's t-test:")
println("  statistic: ${String.format("%.3f", result.statistic)}")
println("  p-value: ${String.format("%.6f", result.pValue)}")
println("  95% CI: [${result.confidenceInterval?.let { "${String.format("%.2f", it.first)}, ${String.format("%.2f", it.second)}" }}]")
println("  Significant: ${result.isSignificant()}")
```

```kotlin
val meanDiff = controlSummary.mean - treatmentSummary.mean
val ciLow = result.confidenceInterval?.first ?: 0.0
val ciHigh = result.confidenceInterval?.second ?: 0.0

plot {
    points {
        x(listOf(meanDiff)); y(listOf("Mean Difference"))
        size = 8.0
        color = Color.hex("#1971c2")
    }
    line {
        x(listOf(ciLow, ciHigh)); y(listOf("Mean Difference", "Mean Difference"))
        color = Color.hex("#1971c2")
        width = 3.0
    }
    line {
        x(listOf(0.0, 0.0)); y(listOf("", "Mean Difference"))
        color = Color.hex("#e03131")
        width = 1.5
    }
    layout { title = "Mean Difference with 95% Confidence Interval"; size = 700 to 300 }
}
```

```kotlin
val equalVar = tTest(controlDurationSec, treatmentDurationSec, equalVariances = true)
println("Equal-variance t-test p=${String.format("%.6f", equalVar.pValue)}")
```

### Non-parametric: Mann-Whitney U

```kotlin
val mwResult = mannWhitneyUTest(controlDurationSec, treatmentDurationSec)
println("Mann-Whitney U: statistic=${String.format("%.1f", mwResult.statistic)}, p=${String.format("%.6f", mwResult.pValue)}")
```

### One-sided test

```kotlin
val oneSided = tTest(controlDurationSec, treatmentDurationSec, alternative = Alternative.GREATER)
println("One-sided (control > treatment) p=${String.format("%.6f", oneSided.pValue)}")
```

```kotlin
val testNames = listOf("Welch t-test", "Equal-var t-test", "Mann-Whitney U", "One-sided t-test")
val testPvalues = listOf(result.pValue, equalVar.pValue, mwResult.pValue, oneSided.pValue)

// p-values are extremely small — use -log10 scale for visibility
val negLog10 = testPvalues.map { -kotlin.math.ln(it) / kotlin.math.ln(10.0) }
val threshold = -kotlin.math.ln(0.05) / kotlin.math.ln(10.0) // -log10(0.05) ≈ 1.3

plot {
    bars {
        x(testNames); y(negLog10)
        fillColor = Color.hex("#9775fa")
        alpha = 0.8
    }
    line {
        x(listOf(testNames.first(), testNames.last())); y(listOf(threshold, threshold))
        color = Color.hex("#e03131")
        width = 2.0
    }
    layout { title = "-log10(p-value) — Higher = More Significant (red = α=0.05)"; size = 800 to 400 }
}
```

## Step 4: Second Metric (Checkout Steps)

```kotlin
val controlSteps = doubleArrayOf(
    3.0, 4.0, 3.0, 5.0, 3.0, 4.0, 4.0, 5.0, 3.0, 4.0,
    4.0, 3.0, 4.0, 3.0, 5.0, 4.0, 4.0, 3.0, 4.0, 5.0
)
val treatmentSteps = doubleArrayOf(
    5.0, 5.0, 4.0, 5.0, 5.0, 5.0, 4.0, 5.0, 5.0, 5.0,
    4.0, 5.0, 5.0, 4.0, 5.0, 5.0, 5.0, 4.0, 5.0, 5.0
)

println("Control steps normality p=${String.format("%.4f", shapiroWilkTest(controlSteps).pValue)}")
val stepsResult = mannWhitneyUTest(controlSteps, treatmentSteps)
println("Mann-Whitney p=${String.format("%.6f", stepsResult.pValue)}, significant=${stepsResult.isSignificant()}")
```

```kotlin
val stepValues = listOf(3, 4, 5)
val controlCounts = stepValues.map { v -> controlSteps.count { it == v.toDouble() } }
val treatmentCounts = stepValues.map { v -> treatmentSteps.count { it == v.toDouble() } }

val allStepX = stepValues.map { "$it steps" } + stepValues.map { "$it steps" }
val allStepY = controlCounts + treatmentCounts
val allStepGroup = stepValues.map { "Control" } + stepValues.map { "Treatment" }

plot {
    bars {
        x(allStepX); y(allStepY.map { it.toDouble() })
        fillColor(allStepGroup)
        alpha = 0.8
    }
    layout { title = "Completed Steps: Control vs Treatment"; size = 700 to 400 }
}
```

## Step 5: Correlation Between Metrics

```kotlin
val correlation = spearmanCorrelation(treatmentDurationSec, treatmentSteps)
println("Spearman ρ = ${String.format("%.3f", correlation.coefficient)}")
println("p-value = ${String.format("%.4f", correlation.pValue)}")
```

```kotlin
// Add jitter to y-axis for discrete step values to avoid overlapping points
val rng = java.util.Random(42)
val jitteredSteps = treatmentSteps.map { it + (rng.nextDouble() - 0.5) * 0.3 }

plot {
    points {
        x(treatmentDurationSec.toList()); y(jitteredSteps)
        color = Color.hex("#e03131")
        size = 6.0
    }
    layout { title = "Treatment: Duration vs Steps (Spearman ρ=${String.format("%.3f", correlation.coefficient)})"; size = 700 to 500 }
}
```

## Paired Before/After Comparison

When the same users are measured before and after a change, use paired tests.

```kotlin
val beforeMs = doubleArrayOf(
    340.2, 415.0, 387.1, 451.3, 369.5, 423.8, 398.0, 446.2, 374.1, 401.5
)
val afterMs = doubleArrayOf(
    310.5, 380.2, 355.8, 410.7, 335.1, 392.4, 365.3, 405.9, 340.8, 371.6
)

val paired = pairedTTest(beforeMs, afterMs)
println("Paired t-test: p=${String.format("%.4f", paired.pValue)}, CI=[${paired.confidenceInterval?.let { "${String.format("%.1f", it.first)}, ${String.format("%.1f", it.second)}" }}]")

val wilcoxon = wilcoxonSignedRankTest(beforeMs, afterMs)
println("Wilcoxon signed-rank: p=${String.format("%.4f", wilcoxon.pValue)}")
```

```kotlin
val users = (1..10).toList()

plot {
    // Connecting lines for each user
    line { x(listOf(1.0, 1.0)); y(listOf(beforeMs[0], afterMs[0])); color = Color.hex("#ced4da"); width = 1.0 }
    line { x(listOf(2.0, 2.0)); y(listOf(beforeMs[1], afterMs[1])); color = Color.hex("#ced4da"); width = 1.0 }
    line { x(listOf(3.0, 3.0)); y(listOf(beforeMs[2], afterMs[2])); color = Color.hex("#ced4da"); width = 1.0 }
    line { x(listOf(4.0, 4.0)); y(listOf(beforeMs[3], afterMs[3])); color = Color.hex("#ced4da"); width = 1.0 }
    line { x(listOf(5.0, 5.0)); y(listOf(beforeMs[4], afterMs[4])); color = Color.hex("#ced4da"); width = 1.0 }
    line { x(listOf(6.0, 6.0)); y(listOf(beforeMs[5], afterMs[5])); color = Color.hex("#ced4da"); width = 1.0 }
    line { x(listOf(7.0, 7.0)); y(listOf(beforeMs[6], afterMs[6])); color = Color.hex("#ced4da"); width = 1.0 }
    line { x(listOf(8.0, 8.0)); y(listOf(beforeMs[7], afterMs[7])); color = Color.hex("#ced4da"); width = 1.0 }
    line { x(listOf(9.0, 9.0)); y(listOf(beforeMs[8], afterMs[8])); color = Color.hex("#ced4da"); width = 1.0 }
    line { x(listOf(10.0, 10.0)); y(listOf(beforeMs[9], afterMs[9])); color = Color.hex("#ced4da"); width = 1.0 }
    // Before points
    points {
        x(users.map { it.toDouble() }); y(beforeMs.toList())
        color = Color.hex("#1971c2")
        size = 6.0
    }
    // After points
    points {
        x(users.map { it.toDouble() }); y(afterMs.toList())
        color = Color.hex("#e03131")
        size = 6.0
    }
    layout { title = "Before (blue) vs After (red) — Connected Pairs"; size = 800 to 450 }
}
```

```kotlin
plot {
    densityPlot(beforeMs.toList()) {
        fillColor = Color.hex("#4dabf7")
        alpha = 0.4
    }
    densityPlot(afterMs.toList()) {
        fillColor = Color.hex("#ff8787")
        alpha = 0.4
    }
    layout { title = "Before (blue) vs After (red): Density Comparison"; size = 800 to 400 }
}
```

```kotlin
val differences = (0 until 10).map { beforeMs[it] - afterMs[it] }
val meanDiffPaired = differences.average()

plot {
    histogram(differences) {
        fillColor = Color.hex("#ffd43b")
        alpha = 0.7
    }
    line {
        x(listOf(meanDiffPaired, meanDiffPaired)); y(listOf(0.0, 4.0))
        color = Color.hex("#1971c2")
        width = 1.5
    }
    layout { title = "Before−After Differences (blue line = mean ${String.format("%.1f", meanDiffPaired)} ms)"; size = 700 to 400 }
}
```

## Summary

The simplified checkout flow (treatment) significantly reduced session duration while increasing completed steps. Both parametric and non-parametric tests confirm the effect.
