# Exploratory Data Analysis with kstats

This notebook walks through a full exploratory data analysis workflow using **kstats** and **Kandy** for visualization.

We will analyze backend service performance metrics collected over 30 days:
- **Response time** (ms)
- **Errors per hour**
- **Memory usage** (MB)
- **Throughput** (requests/second)

The analysis covers descriptive statistics, distribution fitting, correlation, period comparison, normalization, and composite anomaly scoring.

## Dataset

```kotlin
val responseTimeMs = doubleArrayOf(
    89.2, 95.1, 87.6, 102.3, 91.8, 88.4, 96.7, 103.5, 90.1, 94.3,
    88.9, 97.2, 105.8, 91.4, 93.6, 87.1, 99.0, 92.5, 96.1, 104.2,
    90.7, 88.3, 101.6, 93.9, 95.4, 89.8, 98.3, 106.1, 91.0, 94.7
)
val errorsPerHour = doubleArrayOf(
    2.0, 3.0, 1.0, 5.0, 2.0, 1.0, 4.0, 6.0, 2.0, 3.0,
    1.0, 4.0, 7.0, 2.0, 3.0, 1.0, 5.0, 2.0, 4.0, 6.0,
    2.0, 1.0, 5.0, 3.0, 3.0, 1.0, 4.0, 8.0, 2.0, 3.0
)
val memoryUsageMb = doubleArrayOf(
    512.3, 528.1, 505.7, 545.2, 519.6, 508.4, 534.8, 551.3, 515.0, 526.7,
    509.2, 537.1, 558.4, 517.8, 524.3, 503.1, 541.6, 520.9, 531.5, 549.7,
    514.2, 506.8, 543.9, 522.5, 529.0, 511.4, 539.3, 561.2, 516.3, 527.4
)
val throughputRps = doubleArrayOf(
    245.0, 238.0, 251.0, 225.0, 242.0, 249.0, 232.0, 218.0, 244.0, 236.0,
    250.0, 230.0, 212.0, 243.0, 237.0, 253.0, 227.0, 241.0, 233.0, 220.0,
    246.0, 252.0, 224.0, 239.0, 235.0, 248.0, 228.0, 210.0, 243.0, 234.0
)
val days = (1..30).toList()
```

## Step 1: Summary Statistics

```kotlin
val rtSummary = responseTimeMs.describe()
val errSummary = errorsPerHour.describe()
val memSummary = memoryUsageMb.describe()
val tpSummary = throughputRps.describe()

println("Response Time — mean: %.1f, sd: %.1f, min: %.1f, max: %.1f".format(rtSummary.mean, rtSummary.standardDeviation, rtSummary.min, rtSummary.max))
println("Errors/Hour   — mean: %.1f, sd: %.1f, min: %.1f, max: %.1f".format(errSummary.mean, errSummary.standardDeviation, errSummary.min, errSummary.max))
println("Memory (MB)   — mean: %.1f, sd: %.1f, min: %.1f, max: %.1f".format(memSummary.mean, memSummary.standardDeviation, memSummary.min, memSummary.max))
println("Throughput     — mean: %.1f, sd: %.1f, min: %.1f, max: %.1f".format(tpSummary.mean, tpSummary.standardDeviation, tpSummary.min, tpSummary.max))
```

```kotlin
plot {
    line {
        x(days.map { it.toDouble() }); y(responseTimeMs.toList())
        color = Color.hex("#1971c2")
        width = 1.5
    }
    points {
        x(days.map { it.toDouble() }); y(responseTimeMs.toList())
        color = Color.hex("#1971c2")
        size = 3.0
    }
    layout { title = "Response Time (ms) over 30 Days"; size = 800 to 350 }
}
```

```kotlin
plot {
    bars {
        x(days); y(errorsPerHour.toList())
        fillColor = Color.hex("#e03131")
        alpha = 0.7
    }
    layout { title = "Errors per Hour over 30 Days"; size = 800 to 350 }
}
```

```kotlin
plot {
    line {
        x(days.map { it.toDouble() }); y(memoryUsageMb.toList())
        color = Color.hex("#2b8a3e")
        width = 1.5
    }
    points {
        x(days.map { it.toDouble() }); y(memoryUsageMb.toList())
        color = Color.hex("#2b8a3e")
        size = 3.0
    }
    layout { title = "Memory Usage (MB) over 30 Days"; size = 800 to 350 }
}
```

```kotlin
plot {
    line {
        x(days.map { it.toDouble() }); y(throughputRps.toList())
        color = Color.hex("#7048e8")
        width = 1.5
    }
    points {
        x(days.map { it.toDouble() }); y(throughputRps.toList())
        color = Color.hex("#7048e8")
        size = 3.0
    }
    layout { title = "Throughput (req/s) over 30 Days"; size = 800 to 350 }
}
```

## Step 2: Distribution Shape

```kotlin
val metrics = listOf("Response Time", "Errors/Hour", "Memory", "Throughput")
val skewValues = listOf(responseTimeMs.skewness(), errorsPerHour.skewness(), memoryUsageMb.skewness(), throughputRps.skewness())
val kurtValues = listOf(responseTimeMs.kurtosis(), errorsPerHour.kurtosis(), memoryUsageMb.kurtosis(), throughputRps.kurtosis())

metrics.forEachIndexed { i, name ->
    println("$name — skewness: %.3f, kurtosis: %.3f".format(skewValues[i], kurtValues[i]))
}
```

```kotlin
plot {
    bars {
        x(metrics); y(skewValues)
        fillColor = Color.hex("#4dabf7")
        alpha = 0.8
    }
    line {
        x(listOf(metrics.first(), metrics.last())); y(listOf(0.0, 0.0))
        color = Color.hex("#868e96")
        width = 1.0
    }
    layout { title = "Skewness by Metric (0 = symmetric)"; size = 800 to 400 }
}
```

### Normality Tests

```kotlin
val pValues = listOf(
    shapiroWilkTest(responseTimeMs).pValue,
    shapiroWilkTest(errorsPerHour).pValue,
    shapiroWilkTest(memoryUsageMb).pValue,
    shapiroWilkTest(throughputRps).pValue
)
metrics.forEachIndexed { i, name ->
    println("$name — Shapiro-Wilk p=${String.format("%.4f", pValues[i])} ${if (pValues[i] > 0.05) "\u2713 normal" else "\u2717 non-normal"}")
}
```

```kotlin
plot {
    bars {
        x(metrics); y(pValues)
        fillColor = Color.hex("#69db7c")
        alpha = 0.8
    }
    line {
        x(listOf(metrics.first(), metrics.last())); y(listOf(0.05, 0.05))
        color = Color.hex("#e03131")
        width = 2.0
    }
    layout { title = "Shapiro-Wilk p-values (red line = \u03b1=0.05)"; size = 800 to 400 }
}
```

### Fit a Candidate Distribution

```kotlin
val rtFit = NormalDistribution(mu = responseTimeMs.mean(), sigma = responseTimeMs.standardDeviation())
val rtKS = kolmogorovSmirnovTest(responseTimeMs, rtFit)
println("Response Time KS: statistic=${String.format("%.4f", rtKS.statistic)}, p=${String.format("%.4f", rtKS.pValue)}")

val memFit = NormalDistribution(mu = memoryUsageMb.mean(), sigma = memoryUsageMb.standardDeviation())
val memKS = kolmogorovSmirnovTest(memoryUsageMb, memFit)
println("Memory KS: statistic=${String.format("%.4f", memKS.statistic)}, p=${String.format("%.4f", memKS.pValue)}")
```

```kotlin
val xRT = (850..1070).map { it / 10.0 }
val yRTFit = xRT.map { rtFit.pdf(it) }

plot {
    densityPlot(responseTimeMs.toList()) {
        fillColor = Color.hex("#a5d8ff")
        alpha = 0.5
    }
    line {
        x(xRT); y(yRTFit)
        color = Color.hex("#e03131")
        width = 2.0
    }
    layout { title = "Response Time: Kernel Density vs Fitted Normal"; size = 800 to 400 }
}
```

```kotlin
val xMem = (4900..5700).map { it / 10.0 }
val yMemFit = xMem.map { memFit.pdf(it) }

plot {
    densityPlot(memoryUsageMb.toList()) {
        fillColor = Color.hex("#b2f2bb")
        alpha = 0.5
    }
    line {
        x(xMem); y(yMemFit)
        color = Color.hex("#e03131")
        width = 2.0
    }
    layout { title = "Memory Usage: Kernel Density vs Fitted Normal"; size = 800 to 400 }
}
```

## Step 3: Correlations

```kotlin
val matrix = correlationMatrix(responseTimeMs, errorsPerHour, memoryUsageMb, throughputRps)

val labels = listOf("RT", "Errors", "Memory", "Throughput")
for (i in labels.indices) {
    for (j in labels.indices) {
        print("${String.format("%7.3f", matrix[i][j])} ")
    }
    println("  \u2190 ${labels[i]}")
}
```

```kotlin
val heatLabels = listOf("RT", "Errors", "Memory", "Throughput")
val heatX = mutableListOf<String>()
val heatY = mutableListOf<String>()
val heatV = mutableListOf<Double>()
for (i in heatLabels.indices) {
    for (j in heatLabels.indices) {
        heatX.add(heatLabels[j])
        heatY.add(heatLabels[i])
        heatV.add(matrix[i][j])
    }
}

plot {
    tiles {
        x(heatX); y(heatY)
        fillColor(heatV)
    }
    layout { title = "Correlation Matrix"; size = 600 to 500 }
}
```

```kotlin
val rtVsErrors = pearsonCorrelation(responseTimeMs, errorsPerHour)
println("RT vs Errors — Pearson r=${String.format("%.3f", rtVsErrors.coefficient)}, p=${String.format("%.4f", rtVsErrors.pValue)}")

val rtVsThroughput = spearmanCorrelation(responseTimeMs, throughputRps)
println("RT vs Throughput — Spearman \u03c1=${String.format("%.3f", rtVsThroughput.coefficient)}, p=${String.format("%.4f", rtVsThroughput.pValue)}")
```

```kotlin
plot {
    points {
        x(errorsPerHour.toList()); y(responseTimeMs.toList())
        color = Color.hex("#e03131")
        size = 5.0
    }
    layout { title = "Response Time vs Errors (r=${String.format("%.3f", rtVsErrors.coefficient)})"; size = 700 to 500 }
}
```

```kotlin
plot {
    points {
        x(throughputRps.toList()); y(responseTimeMs.toList())
        color = Color.hex("#7048e8")
        size = 5.0
    }
    layout { title = "Response Time vs Throughput (\u03c1=${String.format("%.3f", rtVsThroughput.coefficient)})"; size = 700 to 500 }
}
```

### Regression

```kotlin
val regression = simpleLinearRegression(errorsPerHour, responseTimeMs)

println("Slope: ${String.format("%.2f", regression.slope)} ms per error/hour")
println("Intercept: ${String.format("%.2f", regression.intercept)} ms")
println("R\u00b2: ${String.format("%.3f", regression.rSquared)}")
println("Predicted RT at 4 errors/hour: ${String.format("%.1f", regression.predict(4.0))} ms")
```

## Step 4: Compare Periods

```kotlin
val firstHalfRt = responseTimeMs.sliceArray(0 until 15)
val secondHalfRt = responseTimeMs.sliceArray(15 until 30)

val periodComparison = tTest(firstHalfRt, secondHalfRt)
println("t-test p-value: ${String.format("%.4f", periodComparison.pValue)}")
println("Significant shift: ${periodComparison.isSignificant()}")

val periodRank = mannWhitneyUTest(firstHalfRt, secondHalfRt)
println("Mann-Whitney p-value: ${String.format("%.4f", periodRank.pValue)}")
```

```kotlin
plot {
    densityPlot(firstHalfRt.toList()) {
        fillColor = Color.hex("#4dabf7")
        alpha = 0.4
    }
    densityPlot(secondHalfRt.toList()) {
        fillColor = Color.hex("#ff8787")
        alpha = 0.4
    }
    layout { title = "Response Time: Days 1–15 (blue) vs Days 16–30 (red)"; size = 800 to 400 }
}
```

```kotlin
val firstHalfTp = throughputRps.sliceArray(0 until 15)
val secondHalfTp = throughputRps.sliceArray(15 until 30)

val tpComparison = tTest(firstHalfTp, secondHalfTp)
println("Throughput t-test p-value: ${String.format("%.4f", tpComparison.pValue)}")
```

## Step 5: Normalize and Rank

```kotlin
val rtNormalized = responseTimeMs.zScore()
val memNormalized = memoryUsageMb.zScore()
val rtScaled = responseTimeMs.minMaxNormalize()
val tpScaled = throughputRps.minMaxNormalize()
val rtRanked = responseTimeMs.rank()

println("Z-score sample (RT): ${rtNormalized.take(5).map { String.format("%.2f", it) }}")
println("Min-max sample (RT): ${rtScaled.toList().take(5).map { String.format("%.2f", it) }}")
```

```kotlin
val allDays = days + days
val allZscores = rtNormalized.toList() + memNormalized.toList()
val allMetricLabel = days.map { "Response Time" } + days.map { "Memory Usage" }

plot {
    line {
        x(allDays.map { it.toDouble() }); y(allZscores)
        color(allMetricLabel)
        width = 1.5
    }
    layout { title = "Z-Score Normalized: Response Time vs Memory"; size = 800 to 400 }
}
```

```kotlin
val allScaled = rtScaled.toList() + tpScaled.toList()
val allScaledLabel = days.map { "Response Time" } + days.map { "Throughput" }

plot {
    line {
        x((days + days).map { it.toDouble() }); y(allScaled)
        color(allScaledLabel)
        width = 1.5
    }
    layout { title = "Min-Max Normalized [0,1]: Response Time vs Throughput"; size = 800 to 400 }
}
```

### Composite Anomaly Score

Z-score normalization is useful for combining metrics into a composite score. By averaging the absolute z-scores across all four metrics for each day, we get a single number indicating how "unusual" that day was overall.

```kotlin
val errNormalized = errorsPerHour.zScore()
val tpNormalized = throughputRps.zScore()
// For throughput, high is good, so invert the sign for anomaly scoring
val compositeScore = (0 until 30).map { i ->
    (kotlin.math.abs(rtNormalized[i]) + kotlin.math.abs(errNormalized[i]) + kotlin.math.abs(memNormalized[i]) + kotlin.math.abs(tpNormalized[i])) / 4.0
}

plot {
    bars {
        x(days); y(compositeScore)
        fillColor = Color.hex("#f08c00")
        alpha = 0.8
    }
    line {
        x(listOf(1.0, 30.0)); y(listOf(1.0, 1.0))
        color = Color.hex("#e03131")
        width = 1.5
    }
    layout { title = "Composite Anomaly Score per Day (red = investigation threshold)"; size = 800 to 400 }
}
```
